# Question 3 HW 1 Electronic Devices:
# displaying plots- 
# integrands of 
### $ n=\int_{E_c}^{\infty}N(E)F_{FD}(E,E_F)dE \\ p=\int_{-\infty}^{E_v}N(E)(1-F_{FD}(E,E_F))dE$
# as a function of 
### $-6.5 eV <E< -2.5 eV$

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import cmath
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
T = 2000 #temperature
K_b = 8.61733*(10 ** (-5)) #boltzmann constant 
```

```python
Ev = -5
Ec = -4
E_vals = np.linspace(-6.5, -2.5, 300)
```

## first plot: $ E_{F_1} = -4.5 eV$

```python
Ef1= -4.5
```

$$y_1  vals=N(E)f_{FD}(E,E_{F_1})=\sqrt{(E-E_c) }\frac{1}{1+e^{(E-E_{F_1} )}} $$

```python
y1_vals = [cmath.sqrt(E-Ec) * (1 / (1+cmath.exp((E-Ef1)/(K_b*T))))
           for E in E_vals]
```

$$y_1  vals2=N(E)(1-f_{FD}(E,E_{F_1}))=\sqrt{(E_v-E) }(1-\frac{1}{1+e^{(E-E_{F_1} )}}) $$

```python
y1_vals2 = [cmath.sqrt(Ev-E) * (1-1 / (1+cmath.exp((E-Ef1)/(K_b*T))))
            for E in E_vals]
```

```python
plt.plot(E_vals, y1_vals, 'r-', linewidth=2)
plt.plot(E_vals, y1_vals2, 'b-', linewidth=2)
plt.plot([Ec, Ec], [0, 0.01], 'k--')
plt.plot([Ev, Ev], [0, 0.01], 'k--')
plt.plot([Ef1, Ef1], [0, 0.01], 'k--')
plt.ylabel('N(E) * fFD(E,EF) or N(E)*(1-fFD(E,EF)) [ev]')
plt.xlabel('E [ev]')
red_patch = mpatches.Patch(color='red', label='N(E)*f(E,E_f)')
blue_patch = mpatches.Patch(color='b', label='N(E)*(1-f(E,E_f))')
plt.legend(handles=[red_patch, blue_patch], loc='best',fontsize=8)
```
