## constants

$$N_D=10^{17} cm^{-3}$$
$$\Delta p(x=0)=10^{-5}\cdot N_D=10^{12}$$
$$\Delta p(x=L)=10^{-3}\cdot N_D=10^{14}$$
$$L=0.06 cm$$
$$L_p=0.001 cm$$
$$n_i=10^{10} cm^{-3}$$
$$T=300 K$$
$$E_g=1.12ev$$
$$E_v(x=0)=0ev$$

## equations

$$\frac{d\Delta p}{dt}=D_p\frac{d^2\Delta p}{dx^2}-\mu_p p\frac{d\varepsilon}{dx}-\mu_p \varepsilon\frac{d\Delta p}{dx}+G_L-\frac{\Delta p}{\tau_p}$$

$$0=D_p\frac{d^2\Delta p}{dx^2}-\frac{\Delta p}{\tau_p}\Rightarrow$$

$$\Delta p(x)=C_1\cdot e^{\frac{x}{L_p}}+C_2\cdot e^{-\frac{x}{L_p}}$$

$$Œîùëù(ùë•=0)=10^{12}=C_1+C_2 $$

$$Œîùëù(ùë•=L)=10^{14}=C_1\cdot e^{\frac{0.06}{0.001}}+C_2\cdot  e^{-\frac{0.06}{0.001}} $$



```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
c2=(10**12*np.exp(60)-10**(-14))/(np.exp(60)-np.exp(-60))
c1=10**12-c2

x=np.linspace(0,0.06)
delta_p=c1*np.exp(x/0.001)+c2*np.exp(-x/0.001)
```

```python
fig ,ax=plt.subplots()
ax.plot(x,delta_p)
ax.set_xlabel('$x [cm]$')
ax.set_ylabel('$E [ev]$');
```

$$F_n=KTln(\frac{n}{n_i})+E_i$$
$$F_p=E_i-KTln(\frac{p}{n_i})$$

```python
KT=0.026
Eg=1.12
Ei=Eg/2
N_D=10**17
ni=10**10
p_0=ni**2/N_D
p=p_0+delta_p
n=N_D
F_n=KT*np.log(n/ni)+Ei
F_p=Ei-KT*np.log(p/ni)
fig, ax=plt.subplots(figsize=(8,8))
ax.axhline(Eg,c='r')
ax.axhline(Ei,c='y')
ax.axhline(0,c='k')
ax.axhline(F_n,c='g')
ax.plot(x,F_p)
ax.legend(['$E_c$','$E_i$','$E_v$','$F_n$','$F_p$'],loc='best',bbox_to_anchor=(1.2,0.5))
ax.set_xlabel('$x [cm]$')
ax.set_ylabel('$E [ev]$');
```
