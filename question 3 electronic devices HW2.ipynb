## question 3 Electronic Devices HW2 gaussian simulation:
$C(x,t)=\frac{1}{\sigma\sqrt{2\pi}}\exp^{\frac{-(x-\mu)^2}{2\sigma^2}}$


```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
plt.style.use('seaborn')
from statsmodels.stats.weightstats import DescrStatsW
```

## part 1 , wanted solution-gaussian of:
$ C(x,t) =\frac{1}{\sqrt{4\pi Dt}}\exp^{\frac{-x^2}{2Dt}} \Rightarrow_{D=1}  C(x,t) =\frac{1}{\sqrt{4\pi t}}\exp^{\frac{-x^2}{2t}} $



```python
# constants
NUMBER_OF_BOXES=1000
STEPS=200
MID_PARTICALS=2**40


mid=NUMBER_OF_BOXES//2
boxes=np.zeros(NUMBER_OF_BOXES, np.longlong)
boxes[mid]=MID_PARTICALS

for i in range(0,STEPS):
    idx=np.argwhere(boxes)
    for index in idx:
        temp=boxes[index]
        boxes[index]=0
        boxes[index+1]+=temp//2
        boxes[index-1]+=temp//2
        
fig, ax = plt.subplots(figsize=(10,10))

ax.plot(np.linspace(np.argwhere(boxes)[0],np.argwhere(boxes)[-1],len(np.trim_zeros(boxes))),np.trim_zeros(boxes))
ax.set(title='distribution of the particles to cells, D=1',xlabel='cell number',ylabel='number of particles')
ax.set_ylim(0)

weighted_stats = DescrStatsW( np.linspace(np.argwhere(boxes)[0],np.argwhere(boxes)[-1],len(np.trim_zeros(boxes))), 
                             weights=np.trim_zeros(boxes),
                             ddof=0)

print('\033[1m'+"the standard deviation after the process is :"+'\033[0m', weighted_stats.std[0])
plt.show()

```

## part 2 , wanted solution-gaussian of:
$ C(x,t) =\frac{1}{\sqrt{4\pi Dt}}\exp^{\frac{-x^2}{2Dt}} \Rightarrow_{D=4}  C(x,t) =\frac{1}{\sqrt{16\pi t}}\exp^{\frac{-x^2}{4t}} $


```python
# constants
NUMBER_OF_BOXES=1000
STEPS=200
MID_PARTICALS=2**40


mid=NUMBER_OF_BOXES//2
boxes2=np.zeros(NUMBER_OF_BOXES, np.longlong)
boxes2[mid]=MID_PARTICALS

for i in range(0,STEPS):
    idx=np.argwhere(boxes2)
    for index in idx:
        temp=boxes2[index]
        boxes2[index]=0
        boxes2[index+2]+=temp//2
        boxes2[index-2]+=temp//2
        
        
fig, ax = plt.subplots(figsize=(10,10))

ax.plot(np.linspace(np.argwhere(boxes2)[0],np.argwhere(boxes2)[-1],len(np.trim_zeros(boxes2))),np.trim_zeros(boxes2))
ax.set(title='distribution of the particles to cells, D=4',xlabel='cell number',ylabel='number of particles')
ax.set_ylim(0)

weighted_stats2 = DescrStatsW( np.linspace(np.argwhere(boxes2)[0],np.argwhere(boxes2)[-1],len(np.trim_zeros(boxes2))), 
                             weights=np.trim_zeros(boxes2),
                             ddof=0)

print('\033[1m'+"the standard deviation after the process is :"+'\033[0m', weighted_stats2.std[0])
plt.show()
```
