# Electronic Devices HW5 Q3

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn')
```

## const:

$$N_A=10^{17} [cm^{-3}]$$ 

$$10^{15}\leq N_D \leq 10^{19}$$

## equations:


$$ x_n=[\frac{2K_s\varepsilon_0N_AV_{bi}}{qN_D(N_D+N_A)}]^\frac{1}{2}$$

$$ x_p=[\frac{2K_s\varepsilon_0N_DV_{bi}}{qN_A(N_D+N_A)}]^\frac{1}{2}$$

$$V_{bi}=\frac{KT}{q}ln(\frac{N_AN_D}{ni^2})$$

$$w=x_n+x_p$$

```python
N_D=np.linspace(10**15,10**19)
N_A=10**17
eps=8.85*10**-14
k_s=11.8
q=1.6*10**-19
KT=0.026
ni=10**10

v_bi=KT*np.log((N_A*N_D)/float((ni**2)))


x_n=np.sqrt((2*k_s*eps*N_A*v_bi)/(q*N_D*(N_D+N_A)))

x_p=np.sqrt((2*k_s*eps*N_D*v_bi)/(q*N_A*(N_D+N_A)))

w=x_p+x_n

fig, ax=plt.subplots(figsize=(8,8))

ax.plot(N_D/N_A,x_n*10**4)
ax.plot(N_D/N_A,x_p*10**4)
ax.plot(N_D/N_A,w*10**4,ls='--')
ax.legend(['$x_n$','$x_p$','$w$'],fontsize='16')
ax.set_xlabel('$N_D/N_A$');
ax.set_ylabel('$x[μm]$');

```
